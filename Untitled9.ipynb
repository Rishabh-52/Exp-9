```python
#Rishabh Ghosal
#25070123090
#Exp9:Study of Numpy libary in python


#1.Declaration of numpy
import numpy as np
a=np.array([1,10,20,30,40,50,500])
b=np.array([[1,2,3],[4,5,6]])
print(a)
print(b ,"\n")

#2.Dimenson of array
print(a.ndim)
print(b.ndim ,"\n")

#3.Shape of array
print(a.shape)
print(b.shape ,"\n")

#4.Data type of array
print(a.dtype)
print(b.dtype ,"\n")

#5.Making a zero matrix in python
print(np.zeros((2,3)),"\n")

#6.Making a matrix containing only ones matrix
print(np.ones((3,3)),"\n")

#7.Making a identity matrix
print(np.eye((3)),"\n")

#8.Generating an Array starting from 1 to 10 with a gap of 2
print(np.arange(1,10,2),"\n")

#9.Generate an matrix having equal spaces between
print(np.linspace(0,1,4),"\n")

#10.Value operation of an array
print(a+5)
print(b*2,"\n")

#11.Mean of an array
print(np.mean(a))
print(np.float64(100.0),"\n")

#12.max,min,sum of a an array or matrix
print(np.max(a))
print(np.min(a))
print(np.sum(a))
print(np.int64(1),"\n")
```
